**Load the Dataset**

```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
#basic info
df.info()
```

```python
#Statistical summary
df.describe(include="all")
```

**Idetityfy missing values**

```python
df.isnull().sum()
```

- Based on this function I found the null values such as director,cast,country,date_added,rating,duration

```python
#fill missing 'director' and 'cast' with 'Not Available'
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')

# Drop rows where 'country' or 'date_added' is missing
df = df.dropna(subset=['country', 'date_added'])

# Fill missing 'rating' with most frequent value
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
# Drop or fill missing 'duration'
df['duration'] = df['duration'].fillna('Unknown')
```

```python
# Check if all missing values are handled
print(df.isnull().sum())
```

```python
#count duplicate vaues
df.duplicated().sum()
```

**Standardise text values**

```python
# Strip and lowercase certain fields
df['type'] = df['type'].str.strip().str.lower()
df['country'] = df['country'].str.strip()
df['rating'] = df['rating'].str.strip()
```

**Fix date format**

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

**Rename coulmns to be consistence**

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

**Save cleaned dataset**

```python
df.to_csv("netflix_cleaned.csv", index=False)
```

**Interview questions and answers**

**1. What are missing values and how do you handle them?**
- Missing values occur when no data is stored for a variable in an observation.
#### Handling techniques:
- Remove rows/columns with .dropna()
- Impute missing data using .fillna():
- Mean/median for numerical data
- Mode or placeholder (e.g., "Unknown") for categorical data
- Domain-specific methods, like predictive imputation

**2. How do you treat duplicate records?**
- Duplicate records are identical rows that appear more than once.
#### Handling method in Pandas:
- df.drop_duplicates(inplace=True)

**3. Difference between dropna() and fillna() in Pandas?**
- dropna() removes rows or columns with missing values.
- df.dropna(subset=['column_name'])
- fillna() fills missing values with a specified value.
- df['column_name'].fillna('Unknown', inplace=True)
    
**4. What is outlier treatment and why is it important?**
- Outliers are data points significantly different from others.
#### Treatment methods:
- Remove outliers using IQR or Z-score method
- Cap/floor values (e.g., winsorizing)
- Log transformation or scaling
#### Importance:
- Outliers can distort analysis, affect mean/variance, and reduce model performance.

**5. Explain the process of standardizing data.**
- Standardization ensures consistency across the dataset.
#### It includes:
- Converting text to lowercase
- Removing leading/trailing spaces
- Unifying categorical labels (e.g., "Male" vs "male")

#### Date formatting
- Renaming columns (snake_case, no spaces)

**6. How do you handle inconsistent data formats (e.g., date/time)?**
- Use Pandas to_datetime() for date formats:
- df['date_column'] = pd.to_datetime(df['date_column'], errors='coerce')
- This converts all dates to a standard format and marks invalid ones as NaT.

**7. What are common data cleaning challenges?**
- Inconsistent formats (e.g., "USA" vs "United States")
- Missing or invalid data
- Duplicate entries
- Hidden nulls (like "N/A", "?", "unknown")
- Incorrect data types
    
**8. How can you check data quality?**
- df.info() → Check data types and nulls
- df.isnull().sum() → Count missing values
- df.duplicated().sum() → Check for duplicate rows
- df.value_counts() → Spot inconsistent categorical values
- df.describe() → Review summary stats for anomalies
